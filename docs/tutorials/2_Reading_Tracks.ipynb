# Managing sequence lengths with SeqData
All the regions in the toy bed file we used had the same length of 20 and specified to `fixed_length` argument to be `True`. However, several questions arise in the more general case, where we want to read sequences of arbitrary length from arbitrary BED files.

What if we actually want 50bp sequences instead? This can be simply accomplished by setting the `seq_length` argument to 50

```python
sdata2 = read_bigwig(
    bigwigs=bw_fnames,  # bigwig files
    fasta=fasta_fname,  # reference genome
    seq_name="seq",  # name of resulting xarray variable containing sequences
    cov_name="cov",  # name of resulting xarray variable containing coverage
    bed=bed_fname,  # bed file with regions to extract
    samples=["tangermeme", "tangermeme2"],  # sample names
    out=bw_fname.with_suffix(".zarr"),
    fixed_length=50,  # whether all sequences are the same length and if so, what length
    batch_size=1000,  # number of sequences to load at once
    n_jobs=1,  # number of parallel jobs
    overwrite=True,  # overwrite the output file if it exists
)
sdata2
```

You can see that the sequences are now length 50. Internally, SeqData takes the midpoint of the coordinates provided and extends the start and end by half of the desired sequence length.

With a BED file of variable lengths, the same process occurs to create fixed length sequences

```python
sdata2 = read_bigwig(
    bigwigs=bw_fnames,  # bigwig files
    fasta=fasta_fname,  # reference genome
    seq_name="seq",  # name of resulting xarray variable containing sequences
    cov_name="cov",  # name of resulting xarray variable containing coverage
    bed=bed_fname.with_suffix(".variable.bed"),  # bed file with regions to extract
    samples=["tangermeme", "tangermeme2"],  # sample names
    out=bw_fname.with_suffix(".zarr"),
    fixed_length=50,  # whether all sequences are the same length and if so, what length
    batch_size=1000,  # number of sequences to load at once
    n_jobs=1,  # number of parallel jobs
    overwrite=True,  # overwrite the output file if it exists
)
sdata2
```

As we saw in the Tabular data tutorial, when we specify sequences with different lengths, we can also read in coordiantes corresponding to sequences of different lengths:

# Max jitter
When training ML models on genomic sequence data, it is often useful to introduce some jitter the coordinates of the input regions at different iterations of training. This can help the prevent the model from memorizing the sequences at the specific coordinates in the training data and instead learn more generalizable features.

```python
sdata = read_bigwig(
    bigwigs=[bw_fname],  # bigwig files
    fasta=fasta_fname,  # reference genome
    seq_name="seq",  # name of resulting xarray variable containing sequences
    cov_name="cov",  # name of resulting xarray variable containing coverage
    bed=bed_fname,  # bed file with regions to extract
    samples=["tangermeme"],  # sample names
    max_jitter=128,  # maximum jitter to add to the start of each sequence
    out=bw_fname.with_suffix(".zarr"),
    fixed_length=True,  # whether all sequences are the same length
    batch_size=1000,  # number of sequences to load at once
    n_jobs=1,  # number of parallel jobs
    overwrite=True,  # overwrite the output file if it exists
)
sdata
```

When you use the `max_jitter` argument, the actual sequences loaded in are no longer the original lengths specified in the bed file. Specifically, `max_jitter` is subtracted from the start coordinate and added to the end coordinate. This means that the actual sequence length is `seq_length + (max_jitter*2)`.

```python
sdata.seq.shape
```

Coverage is similarly extended in each direction:

```python
import pysam
import random
import pandas as pd

# Open your reference FASTA file
fasta = pysam.FastaFile(fasta_fname)

regions = pd.read_csv(bed_fname, sep='\t', header=None, names=['chrom', 'start', 'end'])

# Open the BAM file for writing
with pysam.AlignmentFile(fname, 'wb', reference_names=fasta.references, reference_lengths=fasta.lengths) as bamfile:
    # For each region in the BED file
    for _, region in regions.iterrows():
        chrom = region['chrom']
        start = region['start']
        end = region['end']
        num_reads = random.randint(1, 1000)  # Randomly choose number of reads (1-10)
        
        for _ in range(num_reads):
            read_start = random.randint(start, end - 10)  # Random starting point for the read
            read_seq = fasta.fetch(chrom, read_start, read_start + 10)  # Fetch 10 base pairs
            
            # Create a new read
            a = pysam.AlignedSegment()
            a.query_name = "read_" + str(random.randint(1, 10000))
            a.query_sequence = read_seq
            a.flag = 0  # Assume all reads are mapped
            a.reference_id = bamfile.get_tid(chrom)
            a.reference_start = read_start
            a.mapping_quality = 255
            a.cigar = [(0, 10)]  # Match/Mismatch for 10 bases
            
            # Write to BAM
            bamfile.write(a)

# Sort the BAM file
pysam.sort("-o", str(fname.with_suffix(".sorted.bam")), str(fname))

# Index the BAM file
pysam.index(str(fname.with_suffix(".sorted.bam")))
```
