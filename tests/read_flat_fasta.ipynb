# `friedman21`

```python
dataset_dir = "/cellar/users/aklie/data/ml4gland/pubs/friedman21"
```

```python
#
enhancer_fasta = sd.FlatFASTA(
    name="seq",
    fasta=os.path.join(dataset_dir, "ActivityBins/strongEnhancer.fasta"),
    batch_size=1000,
    n_threads=1,
)

silencer_fasta = sd.FlatFASTA(
    name="seq",
    fasta=os.path.join(dataset_dir, "ActivityBins/silencer.fasta"),
    batch_size=1000,
    n_threads=1,
)

white_test_set_fasta = sd.FlatFASTA(
    name="seq",
    fasta=os.path.join(dataset_dir, "white2013TestSet.fasta"),
    batch_size=1000,
    n_threads=1,
)
```

```python
# 
friedman_enhancer_sdata = sd.from_flat_files(
    enhancer_fasta,
    path=os.path.join(dataset_dir, "seqdatas", "strongEnhancer.zarr"),
    fixed_length=False,
    overwrite=True,
)

friedman_silencer_sdata = sd.from_flat_files(
    silencer_fasta,
    path=os.path.join(dataset_dir, "seqdatas", "silencer.zarr"),
    fixed_length=False,
    overwrite=True,
)

friedman_white_test_set_sdata = sd.from_flat_files(
    white_test_set_fasta,
    path=os.path.join(dataset_dir, "seqdatas", "white2013TestSet.zarr"),
    fixed_length=False,
    overwrite=True,
)
```

```python
# Test out loading all into same sdata
friedman_sdata = sd.from_flat_files(
    enhancer_fasta,
    silencer_fasta,
    white_test_set_fasta,
    path=os.path.join(dataset_dir, "seqdatas", "friedman21.zarr"),
    fixed_length=False,
    overwrite=True,
)
friedman_sdata
```
